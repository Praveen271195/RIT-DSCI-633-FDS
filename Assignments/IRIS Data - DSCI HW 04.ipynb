## Standardization of input features

```python
# Importing the standardization lib
from sklearn.preprocessing import StandardScaler
```

```python
# Instance
scaling = StandardScaler()
```

```python
# Columns
iris_data.columns
columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'] 

# Fitting the dataset input features into standardization 
scaled_features = scaling.fit_transform(iris_data[columns])
```

```python
# Casting the scaled features into a dataframe
df_feat= pd.DataFrame(scaled_features, columns=iris_data.columns[:-1])

# Standardized input features
df_feat.head()
```

```python
# Changing the data type of the target feature
target_feature = iris_data["Species"].astype('category')
target_feature.dtypes
```

```python
# Encoding the new target column
target_feature = target_feature.cat.codes
```

```python
# Checking the encoded values
target_feature.value_counts()
```

```python
# Importing metrics to calculate the performance
from sklearn.metrics import classification_report,confusion_matrix
```

```python
# Splitting dataset for training and testing
from sklearn.model_selection import train_test_split
```

```python
# input features (scaled)
X = df_feat #(Using scaled input features)
# output features (categorical - encoded)
y = target_feature #(Using encoded values)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
# Importing linear SVC from SVM
from sklearn.svm import SVC  
```

```python
# Creating instance for SVC model and specifying the parameters, Kernel to linear and gamma to auto
linearSvcModel = SVC(kernel="linear", C=10.0, gamma="auto")
```

```python
# Fitting the data into ML model
linearSvcModel.fit(X_train,y_train) #Fitting model in training features
```

```python
# Saving predictions
linearSvcPredictions = linearSvcModel.predict(X_test)
```

### Finding best error rate and best accuracy manually 

```python
import numpy as np
# Checking the error rate for other K values
error_rate = []
for i in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    prediction_i = knn.predict(X_test)
    error_rate.append(np.mean(prediction_i != y_test))
```

```python
# Plotting the error rate to see the minimum value
plt.plot(range(1,30),error_rate,linestyle = "dashed")
plt.xlabel("k values")
plt.ylabel("Error rate")
plt.title("Error rate vs K value") 
```

**Observation**: We can see that the error is increases when the k value is goes beyond 17

```python
from sklearn import metrics

# Plotting the Knn model acuracy with n values ranging from 1 to 30
k_value = list(range(1,30))
accuracy_scores = [] # Empty list to store the scores 

#Looping through the range specified above
for k in k_value:
    # Instance of model
    knn = KNeighborsClassifier(n_neighbors=k)
    #Fitting the dataset into the Knn algo
    knn.fit(X_train, y_train)
    #Saving prediction values 
    y_predictions = knn.predict(X_test)
    #Computing the scores with y_test and y_predictions
    accuracy_scores.append(metrics.accuracy_score(y_test, y_predictions))

#Plotting the scores wrt to different k values    
plt.plot(k_value, accuracy_scores,linestyle = "dashed")
plt.xlabel('K value')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores Vs k value of KNN model')
plt.show()
```

**Observation**: The accuracy value drops as we the K value goes beyond 17
